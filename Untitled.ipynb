```python
import os

def is_file_in_scope(filename):
    # Check if the file is a .jsx or .css file, or if it matches one of the specific Python filenames
    return (
        filename.endswith('.jsx') or
        filename.endswith('.css') or
        filename in ['serializers.py', 'views.py', 'urls.py', 'models.py', 'settings.py']
    )

def read_files_in_directory(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if is_file_in_scope(file):
                file_path = os.path.join(root, file)
                try:
                    with open(file_path, 'r') as f:
                        print(f"File: {file_path}")
                        content = f.read()
                        print(content)
                        print("-" * 80)  # Separator for readability
                except Exception as e:
                    print(f"Could not read file {file_path}: {e}")
```

```python
if __name__ == "__main__":
    for directory_to_search in ['frontend/src', 'backend']:
        # directory_to_search = input("Enter the directory to search: ")
        read_files_in_directory(directory_to_search)
```
